```python
import pandas as pd 
import numpy as np 
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
baloons_width = np.random.normal(30,1,1000)
baloons_height = np.random.normal(15,1,1000)
baloons_weight = np.random.normal(1,0.5,1000)
melons_width = np.random.normal(25,1,1000)
melons_height = np.random.normal(15,1,1000)
melons_weight = np.random.normal(100,50,1000)

```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.grid()
# ax.plot3D(baloons_width, baloons_height, baloons_weight)
plt.scatter(baloons_width,baloons_height,baloons_weight,  color = 'red', marker="o" ,edgecolors='none')
plt.scatter(melons_width,melons_height,melons_weight, color = 'orange',  marker="^"  , alpha = 0.3 ,edgecolors='none')

ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
ax.set_zlabel('zlabel')

plt.show()
```
